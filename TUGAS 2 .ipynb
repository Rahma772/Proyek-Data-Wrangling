**Mengubah Missing Value**

```python
import numpy as np

# Mengubah simbol missing menjadi NaN
df = df.replace("?", np.nan)
df
```

**Menampilkan Jumlah Data Hilang**

```python
df.isnull().sum()
```

Variabel normalized-losses memiliki jumlah missing value paling banyak yaitu 41, sehingga menjadi prioritas utama dalam proses pembersihan data. Sementara itu, variabel lain memiliki jumlah data hilang yang lebih sedikit dan masih dapat ditangani dengan metode imputasi yang sesuai.

**Mengganti Missing Value pada Variabel Kategori**

```python
# Variabel kategori
kategori = ['make','bahan-bakar','jumlah-pintu','body-style']

# Mengganti missing value dengan "unknown"
df[kategori] = df[kategori].fillna("unknown")
for col in kategori:
    print(f"\nRingkasan kategori {col}:")
    print(df[col].value_counts())
df
```

Secara keseluruhan, data kategori didominasi oleh beberapa kategori utama seperti Toyota (make), gas (bahan bakar), 4 pintu, dan sedan. Nilai "unknown" pada setiap variabel menunjukkan adanya missing value yang telah berhasil ditangani, dengan jumlah yang relatif kecil sehingga tidak terlalu mempengaruhi distribusi data secara keseluruhan.

**Mengganti Missing Value pada Variabel Numerik**

```python
# Variabel numerik
numerik = ['length', 'width', 'height']

# Menghitung mean dan mengganti missing value
for col in numerik:
    mean_value = df[col].mean()
    df[col].fillna(mean_value)

    print(f"Rata-rata {col}:", mean_value)

# Cek apakah masih ada missing value
print("\nJumlah missing value setelah diisi:")
print(df[numerik].isnull().sum())
```

Missing value pada variabel length, width, dan height berhasil diatasi menggunakan rata-rata masing-masing variabel. Setelah proses tersebut, tidak terdapat lagi data yang hilang sehingga dataset siap digunakan untuk tahap analisis atau normalisasi selanjutnya.

**Mengisi Nilai NaN pada variabel Normalised-losses**

Kode di bawah ini digunakan untuk mengubah data pada variabel normalized-losses menjadi tipe numerik agar dapat dilakukan perhitungan statistik.

```python
# Isi NaN dengan mean
df['normalized-losses'] = pd.to_numeric(df['normalized-losses'], errors='coerce')
df  = df.fillna(df.mean(numeric_only=True))
df['normalized-losses'] = df['normalized-losses'].astype(float)
df
```

**Mengisi Nilai NaN pada variabel wheel-bases**

Kode ini bertujuan untuk mengisi nilai kosong pada variabel wheel-bases dengan rata-rata.

```python
# Isi NaN dengan mean
df['wheel-bases'] = pd.to_numeric(df['wheel-bases'], errors='coerce')
df  = df.fillna(df.mean(numeric_only=True))
df['wheel-bases'] = df['wheel-bases'].astype(float)
df
```

**Min-Max Normalization**

Kolom wheel-bases-normalized merupakan hasil normalisasi Min-Max dari variabel wheel-bases. Proses ini dilakukan untuk mengubah skala data ke rentang 0 hingga 1 sehingga memudahkan analisis dan perbandingan antar variabel.

```python
df['wheel-bases-normalized'] = (
    (df['wheel-bases'] - df['wheel-bases'].min()) /
    (df['wheel-bases'].max() - df['wheel-bases'].min())
)

df
```

**Z-Score Normalization**

Hasil Z-score menunjukkan bahwa data telah berhasil dinormalisasi dengan baik, di mana nilai di sekitar 0 menunjukkan data mendekati rata-rata. Nilai negatif menunjukkan data di bawah rata-rata, sedangkan nilai positif menunjukkan data di atas rata-rata. Beberapa nilai ekstrem seperti -2 pada variabel height menunjukkan adanya data yang cukup jauh dari rata-rata.

```python
for col in numerik:
    df[col + '_zscore'] = (df[col] - df[col].mean()) / df[col].std()
df[['length_zscore', 'width_zscore', 'height_zscore']].head()
```
